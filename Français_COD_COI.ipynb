```python
import random

class FrenchPronounQuiz:
    def __init__(self):
        # Sentences with their correct COD/COI forms and type
        # For sentences with multiple valid responses, we use lists
        self.sentences = [
            {
                "original": "Julian connaît ce livre.",
                "correct": "Julian le connaît.",
                "type": "COD",
                "explanation": "Le livre est l'objet direct du verbe connaître, donc on utilise le pronom COD 'le'."
            },
            {
                "original": "Mathieu parle à son ami.",
                "correct": "Mathieu lui parle.",
                "type": "COI",
                "explanation": "Son ami est l'objet indirect (parler à quelqu'un), donc on utilise le pronom COI 'lui'."
            },
            {
                "original": "Marie regarde ses filles.",
                "correct": "Marie les regarde.",
                "type": "COD",
                "explanation": "Ses filles est l'objet direct du verbe regarder, donc on utilise le pronom COD 'les'."
            },
            {
                "original": "Nous donnons des cadeaux à nos parents.",
                "correct": "Nous leur donnons des cadeaux.",
                "type": "COI",
                "explanation": "Nos parents est l'objet indirect (donner à quelqu'un), donc on utilise le pronom COI 'leur'."
            },
            {
                "original": "Elle mange une pomme.",
                "correct": "Elle la mange.",
                "type": "COD",
                "explanation": "Une pomme est l'objet direct du verbe manger, donc on utilise le pronom COD 'la'."
            },
            {
                "original": "Vous écrivez une lettre à votre professeur.",
                "correct": ["Vous lui écrivez une lettre.", "Vous la lui écrivez."],
                "type": "COI/COD",
                "explanation": "Votre professeur est l'objet indirect (écrire à quelqu'un), donc on utilise le pronom COI 'lui'. On peut aussi remplacer 'une lettre' par le COD 'la'."
            },
            {
                "original": "Pierre achète des fleurs pour sa mère.",
                "correct": "Pierre les achète pour sa mère.",
                "type": "COD",
                "explanation": "Des fleurs est l'objet direct du verbe acheter, donc on utilise le pronom COD 'les'."
            },
            {
                "original": "Nous téléphonons à nos amis.",
                "correct": "Nous leur téléphonons.",
                "type": "COI",
                "explanation": "Nos amis est l'objet indirect (téléphoner à quelqu'un), donc on utilise le pronom COI 'leur'."
            },
            {
                "original": "Tu prépares le dîner.",
                "correct": "Tu le prépares.",
                "type": "COD",
                "explanation": "Le dîner est l'objet direct du verbe préparer, donc on utilise le pronom COD 'le'."
            },
            {
                "original": "Je montre les photos à mes grands-parents.",
                "correct": ["Je leur montre les photos.", "Je les leur montre."],
                "type": "COI/COD",
                "explanation": "Mes grands-parents est l'objet indirect (montrer à quelqu'un), donc on utilise le pronom COI 'leur'. On peut aussi remplacer 'les photos' par le COD 'les'."
            },
            {
                "original": "Nous envoyons un message à notre professeur.",
                "correct": ["Nous lui envoyons un message.", "Nous le lui envoyons."],
                "type": "COI/COD",
                "explanation": "Notre professeur est l'objet indirect (envoyer à quelqu'un), donc on utilise le pronom COI 'lui'. On peut aussi remplacer 'un message' par le COD 'le'."
            },
            {
                "original": "Les enfants écoutent la musique.",
                "correct": ["Les enfants l'écoutent.", "Ils l'écoutent."],
                "type": "COD",
                "explanation": "La musique est l'objet direct du verbe écouter, donc on utilise le pronom COD 'l'' (devant une voyelle)."
            },
            {
                "original": "Anne offre des chocolats à son père.",
                "correct": ["Anne lui offre des chocolats.", "Elle lui offre des chocolats.", "Anne les lui offre.", "Elle les lui offre."],
                "type": "COI/COD",
                "explanation": "Son père est l'objet indirect (offrir à quelqu'un), donc on utilise le pronom COI 'lui'. On peut aussi remplacer 'des chocolats' par le COD 'les'."
            },
            {
                "original": "Thomas explique l'exercice à ses élèves.",
                "correct": ["Thomas leur explique l'exercice.", "Il leur explique l'exercice.", "Thomas le leur explique.", "Il le leur explique."],
                "type": "COI/COD",
                "explanation": "Ses élèves est l'objet indirect (expliquer à quelqu'un), donc on utilise le pronom COI 'leur'. On peut aussi remplacer 'l'exercice' par le COD 'le'."
            },
            {
                "original": "Nous avons vu Sophie au cinéma.",
                "correct": ["Nous l'avons vue au cinéma.", "Nous l'avons vue."],
                "type": "COD",
                "explanation": "Sophie est l'objet direct du verbe voir, donc on utilise le pronom COD 'l''."
            }
        ]

    def check_answer(self, original_sentence, user_answer, correct_answers):
        """Check if the user's answer matches any of the correct answers or acceptable variations."""
        user_answer = user_answer.strip()

        # Handle exact matches first
        if isinstance(correct_answers, list):
            if user_answer in correct_answers:
                return True
        elif user_answer == correct_answers:
            return True

        # Handle common variations and minor mistakes

        # Case 1: Missing elision (e.g., "la écoutent" instead of "l'écoutent")
        if "la é" in user_answer:
            fixed_answer = user_answer.replace("la é", "l'é")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "elision"
            elif fixed_answer == correct_answers:
                return "elision"

        # Case 2: Missing elision with "le" (e.g., "le explique" instead of "l'explique")
        if "le e" in user_answer or "le é" in user_answer or "le a" in user_answer:
            fixed_answer = user_answer.replace("le e", "l'e").replace("le é", "l'é").replace("le a", "l'a")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "elision"
            elif fixed_answer == correct_answers:
                return "elision"

        # Case 3: Space after apostrophe (e.g., "l' écoutent" instead of "l'écoutent")
        if "l' " in user_answer:
            fixed_answer = user_answer.replace("l' ", "l'")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "space"
            elif fixed_answer == correct_answers:
                return "space"

        # Case 4: Wrong gender but correct type (e.g., "le pomme" instead of "la pomme")
        if " le " in user_answer and " la " in correct_answers:
            fixed_answer = user_answer.replace(" le ", " la ")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "gender"
            elif fixed_answer == correct_answers:
                return "gender"

        if " la " in user_answer and " le " in correct_answers:
            fixed_answer = user_answer.replace(" la ", " le ")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "gender"
            elif fixed_answer == correct_answers:
                return "gender"

        # Not a match or acceptable variation
        return False

    def explain_answer(self, sentence):
        """Provide explanation for the answer."""
        return sentence["explanation"]

    def run_quiz(self):
        """Run the quiz with random sentences."""
        print("\n=== QUIZ SUR LES PRONOMS COD ET COI EN FRANÇAIS ===")
        print("Transformez la phrase en utilisant un pronom COD ou COI approprié.\n")

        # Shuffle the sentences
        random_sentences = random.sample(self.sentences, min(5, len(self.sentences)))

        score = 0
        total_questions = len(random_sentences)

        for i, sentence in enumerate(random_sentences, 1):
            original = sentence["original"]
            correct = sentence["correct"]

            print(f"\nQuestion {i}/{total_questions}:")
            print(f">> {original} << Comment pouvons-nous écrire cela en utilisant COI ou COD ?")

            user_answer = input("Votre Réponse: ")

            check_result = self.check_answer(original, user_answer, correct)

            if check_result is True:
                print("CORRECTE ! ✓")
                print(self.explain_answer(sentence))
                score += 1
            elif check_result == "elision":
                print("PRESQUE CORRECTE ! La réponse est conceptuellement correcte, mais il faut utiliser l'élision.")
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"La forme correcte serait: {correct_str}")
                else:
                    print(f'La forme correcte serait: "{correct}"')
                print(self.explain_answer(sentence))
                score += 0.5  # Give partial credit
            elif check_result == "space":
                print("PRESQUE CORRECTE ! La réponse est conceptuellement correcte, mais il ne faut pas d'espace après l'apostrophe.")
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"La forme correcte serait: {correct_str}")
                else:
                    print(f'La forme correcte serait: "{correct}"')
                print(self.explain_answer(sentence))
                score += 0.5  # Give partial credit
            elif check_result == "gender":
                print("PRESQUE CORRECTE ! La réponse est conceptuellement correcte, mais le genre du pronom n'est pas correct.")
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"La forme correcte serait: {correct_str}")
                else:
                    print(f'La forme correcte serait: "{correct}"')
                print(self.explain_answer(sentence))
                score += 0.5  # Give partial credit
            else:
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"INCORRECTE ! ✗ La réponse correcte serait: {correct_str}")
                else:
                    print(f'INCORRECTE ! ✗ La réponse correcte serait: "{correct}"')
                print(self.explain_answer(sentence))

        # Print final score - now supporting partial points
        print(f"\nVotre score final: {score}/{total_questions}")
        percentage = (score / total_questions) * 100
        print(f"Pourcentage: {percentage:.1f}%")

        if percentage >= 90:
            print("Excellent ! Vous maîtrisez très bien les pronoms COD et COI !")
        elif percentage >= 70:
            print("Bon travail ! Continuez à pratiquer !")
        elif percentage >= 50:
            print("Pas mal ! Avec un peu plus de pratique, vous allez vous améliorer !")
        else:
            print("Continuez à pratiquer ! Les pronoms COD et COI deviennent plus faciles avec l'exercice.")

        print("\nVoici un petit rappel :")
        print("COD (Complément d'Objet Direct) - répond à la question 'qui ?' ou 'quoi ?'")
        print("   Pronoms COD : le, la, l', les")
        print("COI (Complément d'Objet Indirect) - répond à la question 'à qui ?' ou 'à quoi ?'")
        print("   Pronoms COI : lui, leur")

        play_again = input("\nVoulez-vous jouer à nouveau ? (oui/non): ")
        if play_again.lower() in ["oui", "o", "yes", "y"]:
            self.run_quiz()
        else:
            print("Merci d'avoir joué ! À bientôt !")

# Start the quiz
if __name__ == "__main__":
    quiz = FrenchPronounQuiz()
    quiz.run_quiz()
```
