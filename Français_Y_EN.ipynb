```python
import random

class FrenchYEnQuiz:
    def __init__(self):
        # Sentences with their correct forms using 'y' or 'en' and explanation
        # For sentences with multiple valid responses, we use lists
        self.sentences = [
            {
                "original": "Je vais à Paris.",
                "correct": "J'y vais.",
                "type": "Y",
                "explanation": "'Y' remplace 'à Paris', un complément de lieu introduit par 'à'."
            },
            {
                "original": "Elle habite en France.",
                "correct": "Elle y habite.",
                "type": "Y",
                "explanation": "'Y' remplace 'en France', un complément de lieu."
            },
            {
                "original": "Nous pensons à nos vacances.",
                "correct": "Nous y pensons.",
                "type": "Y",
                "explanation": "'Y' remplace 'à nos vacances', un complément introduit par 'à'."
            },
            {
                "original": "Il a besoin d'aide.",
                "correct": "Il en a besoin.",
                "type": "EN",
                "explanation": "'En' remplace 'd'aide', un complément introduit par 'de'."
            },
            {
                "original": "J'ai mangé trois pommes.",
                "correct": "J'en ai mangé trois.",
                "type": "EN",
                "explanation": "'En' remplace 'pommes' et on garde le nombre 'trois'."
            },
            {
                "original": "Elle parle de son voyage.",
                "correct": "Elle en parle.",
                "type": "EN",
                "explanation": "'En' remplace 'de son voyage', un complément introduit par 'de'."
            },
            {
                "original": "Tu te souviens de cette histoire ?",
                "correct": ["Tu t'en souviens ?", "T'en souviens-tu ?"],
                "type": "EN",
                "explanation": "'En' remplace 'de cette histoire', un complément introduit par 'de'."
            },
            {
                "original": "Nous allons au cinéma.",
                "correct": "Nous y allons.",
                "type": "Y",
                "explanation": "'Y' remplace 'au cinéma', un complément de lieu (à + le cinéma)."
            },
            {
                "original": "Elle a acheté beaucoup de livres.",
                "correct": "Elle en a acheté beaucoup.",
                "type": "EN",
                "explanation": "'En' remplace 'livres' et on garde la quantité 'beaucoup'."
            },
            {
                "original": "Vous réfléchissez à ce problème.",
                "correct": "Vous y réfléchissez.",
                "type": "Y",
                "explanation": "'Y' remplace 'à ce problème', un complément introduit par 'à'."
            },
            {
                "original": "Je viens de la boulangerie.",
                "correct": "J'en viens.",
                "type": "EN",
                "explanation": "'En' remplace 'de la boulangerie', un complément de lieu introduit par 'de'."
            },
            {
                "original": "Pierre a deux frères.",
                "correct": "Pierre en a deux.",
                "type": "EN",
                "explanation": "'En' remplace 'frères' et on garde le nombre 'deux'."
            },
            {
                "original": "Nous participons à la réunion.",
                "correct": "Nous y participons.",
                "type": "Y",
                "explanation": "'Y' remplace 'à la réunion', un complément introduit par 'à'."
            },
            {
                "original": "Tu as besoin de combien d'argent ?",
                "correct": ["Tu en as besoin de combien ?", "De combien en as-tu besoin ?"],
                "type": "EN",
                "explanation": "'En' remplace 'd'argent' et on garde l'interrogation sur la quantité."
            },
            {
                "original": "Elle s'intéresse à la politique.",
                "correct": "Elle s'y intéresse.",
                "type": "Y",
                "explanation": "'Y' remplace 'à la politique', un complément introduit par 'à'."
            },
            {
                "original": "Nous avons pris trois tasses de café.",
                "correct": "Nous en avons pris trois.",
                "type": "EN",
                "explanation": "'En' remplace 'tasses de café' et on garde le nombre 'trois'."
            },
            {
                "original": "Ils ont parlé de leurs projets.",
                "correct": "Ils en ont parlé.",
                "type": "EN",
                "explanation": "'En' remplace 'de leurs projets', un complément introduit par 'de'."
            },
            {
                "original": "Je réponds au message.",
                "correct": "J'y réponds.",
                "type": "Y",
                "explanation": "'Y' remplace 'au message', un complément introduit par 'à' (à + le)."
            },
            {
                "original": "Elle rêve de vacances.",
                "correct": "Elle en rêve.",
                "type": "EN",
                "explanation": "'En' remplace 'de vacances', un complément introduit par 'de'."
            },
            {
                "original": "Je me suis habitué à ce climat.",
                "correct": "Je m'y suis habitué.",
                "type": "Y",
                "explanation": "'Y' remplace 'à ce climat', un complément introduit par 'à'."
            }
        ]

    def check_answer(self, original_sentence, user_answer, correct_answers):
        """Check if the user's answer matches any of the correct answers or acceptable variations."""
        user_answer = user_answer.strip()

        # Handle exact matches first
        if isinstance(correct_answers, list):
            if user_answer in correct_answers:
                return True
        elif user_answer == correct_answers:
            return True

        # Handle common variations and minor mistakes

        # Case 1: Missing elision (e.g., "Je y vais" instead of "J'y vais")
        if "Je y " in user_answer:
            fixed_answer = user_answer.replace("Je y ", "J'y ")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "elision"
            elif fixed_answer == correct_answers:
                return "elision"

        # Case 2: Space after apostrophe (e.g., "J' y vais" instead of "J'y vais")
        if "J' y" in user_answer:
            fixed_answer = user_answer.replace("J' y", "J'y")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "space"
            elif fixed_answer == correct_answers:
                return "space"

        # Case 3: Missing apostrophe in reflexive (e.g., "Tu te en souviens" instead of "Tu t'en souviens")
        if "te en" in user_answer:
            fixed_answer = user_answer.replace("te en", "t'en")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "elision"
            elif fixed_answer == correct_answers:
                return "elision"

        # Case 4: Space after apostrophe in reflexive (e.g., "t' en souviens" instead of "t'en souviens")
        if "t' en" in user_answer:
            fixed_answer = user_answer.replace("t' en", "t'en")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "space"
            elif fixed_answer == correct_answers:
                return "space"

        # Case 5: Missing apostrophe in reflexive (e.g., "Je me y suis habitué" instead of "Je m'y suis habitué")
        if "me y" in user_answer:
            fixed_answer = user_answer.replace("me y", "m'y")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "elision"
            elif fixed_answer == correct_answers:
                return "elision"

        # Case 6: Space after apostrophe in reflexive (e.g., "m' y suis" instead of "m'y suis")
        if "m' y" in user_answer:
            fixed_answer = user_answer.replace("m' y", "m'y")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "space"
            elif fixed_answer == correct_answers:
                return "space"

        # Case 7: Wrong pronoun type but correct usage (e.g., "J'en vais" instead of "J'y vais")
        if " y " in correct_answers and " en " in user_answer:
            fixed_answer = user_answer.replace(" en ", " y ")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "pronoun"
            elif fixed_answer == correct_answers:
                return "pronoun"

        if " en " in correct_answers and " y " in user_answer:
            fixed_answer = user_answer.replace(" y ", " en ")
            if isinstance(correct_answers, list):
                if fixed_answer in correct_answers:
                    return "pronoun"
            elif fixed_answer == correct_answers:
                return "pronoun"

        # Not a match or acceptable variation
        return False

    def explain_answer(self, sentence):
        """Provide explanation for the answer."""
        return sentence["explanation"]

    def run_quiz(self):
        """Run the quiz with random sentences."""
        print("\n=== QUIZ SUR LES PRONOMS Y ET EN EN FRANÇAIS ===")
        print("Transformez la phrase en utilisant les pronoms Y ou EN.\n")

        # Shuffle the sentences
        random_sentences = random.sample(self.sentences, min(5, len(self.sentences)))

        score = 0
        total_questions = len(random_sentences)

        for i, sentence in enumerate(random_sentences, 1):
            original = sentence["original"]
            correct = sentence["correct"]

            print(f"\nQuestion {i}/{total_questions}:")
            print(f">> {original} << Comment pouvons-nous écrire cela en utilisant Y ou EN ?")

            user_answer = input("Votre Réponse: ")

            check_result = self.check_answer(original, user_answer, correct)

            if check_result is True:
                print("CORRECTE ! ✓")
                print(self.explain_answer(sentence))
                score += 1
            elif check_result == "elision":
                print("PRESQUE CORRECTE ! La réponse est conceptuellement correcte, mais il faut utiliser l'élision.")
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"La forme correcte serait: {correct_str}")
                else:
                    print(f'La forme correcte serait: "{correct}"')
                print(self.explain_answer(sentence))
                score += 0.5  # Give partial credit
            elif check_result == "space":
                print("PRESQUE CORRECTE ! La réponse est conceptuellement correcte, mais il ne faut pas d'espace après l'apostrophe.")
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"La forme correcte serait: {correct_str}")
                else:
                    print(f'La forme correcte serait: "{correct}"')
                print(self.explain_answer(sentence))
                score += 0.5  # Give partial credit
            elif check_result == "pronoun":
                print("PRESQUE CORRECTE ! La structure est bonne, mais vous avez utilisé le mauvais pronom (y/en).")
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"La forme correcte serait: {correct_str}")
                else:
                    print(f'La forme correcte serait: "{correct}"')
                print(self.explain_answer(sentence))
                score += 0.5  # Give partial credit
            else:
                if isinstance(correct, list):
                    correct_str = " ou ".join([f'"{c}"' for c in correct])
                    print(f"INCORRECTE ! ✗ La réponse correcte serait: {correct_str}")
                else:
                    print(f'INCORRECTE ! ✗ La réponse correcte serait: "{correct}"')
                print(self.explain_answer(sentence))

        # Print final score - now supporting partial points
        print(f"\nVotre score final: {score}/{total_questions}")
        percentage = (score / total_questions) * 100
        print(f"Pourcentage: {percentage:.1f}%")

        if percentage >= 90:
            print("Excellent ! Vous maîtrisez très bien les pronoms Y et EN !")
        elif percentage >= 70:
            print("Bon travail ! Continuez à pratiquer !")
        elif percentage >= 50:
            print("Pas mal ! Avec un peu plus de pratique, vous allez vous améliorer !")
        else:
            print("Continuez à pratiquer ! Les pronoms Y et EN deviennent plus faciles avec l'exercice.")

        print("\nVoici un petit rappel :")
        print("'Y' remplace :")
        print("   - Un lieu introduit par 'à', 'au', 'à la', 'aux', 'en', 'dans', 'sur', etc.")
        print("   - Un complément introduit par 'à'")
        print("'EN' remplace :")
        print("   - Un complément introduit par 'de'")
        print("   - Une quantité (avec ou sans nombre)")
        print("   - Un lieu introduit par 'de'")

        play_again = input("\nVoulez-vous jouer à nouveau ? (oui/non): ")
        if play_again.lower() in ["oui", "o", "yes", "y"]:
            self.run_quiz()
        else:
            print("Merci d'avoir joué ! À bientôt !")

# Start the quiz
if __name__ == "__main__":
    quiz = FrenchYEnQuiz()
    quiz.run_quiz()
```
